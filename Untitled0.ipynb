```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np

print("TensorFlow version:", tf.__version__)
print("GPU Ğ´Ğ¾ÑÑ‚ÑƒĞ¿ĞµĞ½:", tf.config.list_physical_devices('GPU') is not None)
```

```python
dataset = tfds.load('cats_vs_dogs', split='train', as_supervised=True)

def preprocess(image, label):
    image = tf.image.resize(image, (160, 160))
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

dataset = dataset.map(preprocess)

dataset_size = 23262
train_size = int(0.8 * dataset_size)

train_dataset = dataset.take(train_size)
val_dataset = dataset.skip(train_size)

BATCH_SIZE = 32
train_batches = train_dataset.batch(BATCH_SIZE).prefetch(1)
val_batches = val_dataset.batch(BATCH_SIZE).prefetch(1)

print(f"Ğ¢Ñ€ĞµĞ½Ğ¸Ñ€Ğ¾Ğ²Ğ¾Ñ‡Ğ½Ñ‹Ñ… Ğ±Ğ°Ñ‚Ñ‡ĞµĞ¹: {len(list(train_batches))}")
print(f"Ğ’Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ğ¾Ğ½Ğ½Ñ‹Ñ… Ğ±Ğ°Ñ‚Ñ‡ĞµĞ¹: {len(list(val_batches))}")
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models

base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(160, 160, 3)
)

base_model.trainable = False
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.2),
    layers.Dense(2, activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_batches,
    validation_data=val_batches,
    epochs=5
)
```

```python
# Ğ“Ñ€Ğ°Ñ„Ğ¸Ğº Ñ‚Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚Ğ¸
plt.plot(history.history['accuracy'], label='Ğ¢Ñ€ĞµĞ½Ğ¸Ñ€Ğ¾Ğ²ĞºĞ°')
plt.plot(history.history['val_accuracy'], label='Ğ’Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ñ')
plt.title('Ğ¢Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸')
plt.xlabel('Ğ­Ğ¿Ğ¾Ñ…Ğ°')
plt.ylabel('Ğ¢Ğ¾Ñ‡Ğ½Ğ¾ÑÑ‚ÑŒ')
plt.legend()
plt.show()

# Ğ“Ñ€Ğ°Ñ„Ğ¸Ğº Ğ¿Ğ¾Ñ‚ĞµÑ€ÑŒ
plt.plot(history.history['loss'], label='Ğ¢Ñ€ĞµĞ½Ğ¸Ñ€Ğ¾Ğ²ĞºĞ°')
plt.plot(history.history['val_loss'], label='Ğ’Ğ°Ğ»Ğ¸Ğ´Ğ°Ñ†Ğ¸Ñ')
plt.title('ĞŸĞ¾Ñ‚ĞµÑ€Ğ¸ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸')
plt.xlabel('Ğ­Ğ¿Ğ¾Ñ…Ğ°')
plt.ylabel('ĞŸĞ¾Ñ‚ĞµÑ€Ğ¸')
plt.legend()
plt.show()
```
